## Exploring the parameters in Theorem 1

```python
parameters = np.zeros((10, 9))
for i in range(10):
    items = synthetic_data.data_generation(100, 10)
    pm = synthetic_data.synthetic_pairs(items)
    p = [pm.get_parameters(j) for j in range(1,11)]
    parameters += np.vstack(p)
parameters /= 10
```

## Analyzing inconsistencies

```python
rate = np.zeros((10, 4))
for i in range(10):
    items = synthetic_data.data_generation(100, 10)
    pm = synthetic_data.synthetic_pairs(items)
    r = [pm.inconsistency_analysis(j) for j in range(1,11)]
    rate += np.vstack(r)
rate /= 10
```

```python
plt.plot(x, rate[:,0], 'ro--', label='weak violations')
plt.plot(x, rate[:,1], 'bo--', label='moderate violations')
plt.plot(x, rate[:,2], 'go--', label='strong violations')
plt.plot(x, rate[:,3], 'yo--', label='pairwise inconsistencies')
plt.ylabel('Violations of Rational Choice')
plt.legend()
```

## Estimation Error

```python
error = np.zeros((10, 2))
for i in range(10):
    items = synthetic_data.data_generation(100, 10)
    pm = synthetic_data.synthetic_pairs(items)
    e = [pm.estimation_error(j, 1) for j in range(1,11)]
    error += np.vstack(e)
error /= 10
```
